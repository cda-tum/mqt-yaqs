Define parameters

```python
max_bond = 256

J = 1
g = 1
dt = 0.1
num_qubits = 16
```

```python
# results = generate_heisenberg_observable_data(num_qubits, J, g, dt, pad, thresholds, max_bonds, timesteps_list, min_bond)
min_bond = 2
timesteps_list = [*range(1, 21)]
results = generate_sim_data(
    make_circ=create_heisenberg_circuit,
    make_args=(num_qubits, J, J, J, g, dt),
    timesteps=timesteps_list,
    min_bond_dim=min_bond,
    periodic=False,
    break_on_exceed=True,
    bond_dim_limit=max_bond
)
```
